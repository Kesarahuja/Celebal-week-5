```python
# Check missing values
missing = train.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# Example
train['PoolQC'] = train['PoolQC'].fillna('None')
train['LotFrontage'] = train.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

```python
train = train.drop(['Id'], axis=1)
test_ID = test['Id']
test = test.drop(['Id'], axis=1)
```

```python
qual_mapping = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'None': 0}
train['ExterQual'] = train['ExterQual'].map(qual_mapping)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)

# Align columns of train and test
train, test = train.align(test, join='left', axis=1)
test = test.fillna(0)
```

```python
# Total square footage
train['TotalSF'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']
test['TotalSF'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']

# Total number of bathrooms
train['TotalBath'] = (train['FullBath'] + 0.5 * train['HalfBath'] +
                      train['BsmtFullBath'] + 0.5 * train['BsmtHalfBath'])

test['TotalBath'] = (test['FullBath'] + 0.5 * test['HalfBath'] +
                     test['BsmtFullBath'] + 0.5 * test['BsmtHalfBath'])

# Age of the house
train['Age'] = train['YrSold'] - train['YearBuilt']
test['Age'] = test['YrSold'] - test['YearBuilt']
```

```python
from scipy.stats import skew
import numpy as np # Import numpy

# Select only numerical features (excluding boolean)
skewed_feats = train.select_dtypes(include=np.number).columns
skewness = train[skewed_feats].apply(lambda x: skew(x.dropna()))
skewness = skewness[abs(skewness) > 0.75]

# Apply log1p transformation
for feat in skewness.index:
    train[feat] = np.log1p(train[feat])
    # Check if the feature exists in the test set before applying transformation
    if feat in test.columns:
        test[feat] = np.log1p(test[feat])
```
